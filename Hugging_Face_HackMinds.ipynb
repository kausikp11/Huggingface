```python
from byaldi import RAGMultiModalModel
from huggingface_hub import InferenceClient
from kaggle_secrets import UserSecretsClient
from pdf2image import convert_from_path
from PIL import Image
from loadimg import load_img
```

```python
file_name = ""
text_query = "Please extract the Fiscal year end date and year"
```

```python
images = convert_from_path(file_name)
images[2]
```

```python
RAG = RAGMultiModalModel.from_pretrained("vidore/colpali")
```

```python
RAG.index(
    input_path=file_name,
    index_name="test", # index will be saved at index_root/index_name/
    store_collection_with_index=False,
    overwrite=False
)
```

```python
results = RAG.search(text_query, k=1)
results
```

```python
user_secrets = UserSecretsClient()
secret_value_0 = user_secrets.get_secret("HF_Key")


model = InferenceClient(api_key=secret_value_0)
```

```python
image_index = results[0]["page_num"] - 1
if 0 <= image_index < len(images):
    selected_image = images[image_index]
else:
    raise ValueError("Image index out of range.")
selected_image.save("test.png","PNG")


image_index = results[0]["page_num"] - 1
if 0 <= image_index < len(images):
    selected_image = images[image_index]
else:
    raise ValueError("Image index out of range.")
selected_image.save("test.png","PNG")

# Open the image
image = Image.open("test.png")  # Replace with your image file path

# Reduce the resolution by resizing
new_width = image.width//4  # Set the new width
aspect_ratio = image.height / image.width
new_height = int(new_width * aspect_ratio)
resized_image = image.resize((new_width, new_height), Image.Resampling.LANCZOS)

# Save the resized image
resized_image.save("test_re.png", format="PNG")  # Save with a new name
print("Image resized and saved.")
```

```python
my = load_img('test_re.png',output_type="base64")
# Create the messages object
messages = [
    {
        "role": "user",
        "content": [
            {"type": "image_url", "image_url":{
                "url":my}
             },  # Base64-encoded image
            {"type": "text", "text": text_query},               # Text query
        ],
    }
]

generated_ids = model.chat.completions.create(
    model="Qwen/Qwen2-VL-7B-Instruct",
    messages=messages,
    max_tokens=100
)
#print("Generated Response:", generated_ids)
print(generated_ids.choices[0].message.content)
```
